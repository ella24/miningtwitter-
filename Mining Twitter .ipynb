```python
#The issue while opening the json file was Trailing Data. Solved by lines=True 
df= pd.read_json("beatriz.json", encoding = 'utf8', lines=True)
df.head()
```

```python
df.columns
```

```python
#This is a summary of some tweets with geolocation 
df.place.value_counts()
```

```python
#We need to break the date column into year/month/ect to get the tweets since Rossello resign 
df['month_year'] = pd.to_datetime(df['created_at']).dt.to_period('M')
df.head()
```

```python
df.source.value_counts()
```

```python
# filter rows for year 2019/2020 using  the boolean expression
df2 = df[df['month_year']>='2019-7']
print(df2.shape)
```

```python
df2['month_year'].value_counts()
```

```python
#Using the variable "place"
#we have 15 tweets without the location 
df2.place.isnull().value_counts()
```

```python
# creating bool series False for NaN values  
bool_series = pd.notnull(df2["place"])
bool_series.head()
```

```python
#All the tweets with location 
df2[bool_series]
```

```python
df2[bool_series].to_csv('beatrizwithlocation.csv', index=False)
```
